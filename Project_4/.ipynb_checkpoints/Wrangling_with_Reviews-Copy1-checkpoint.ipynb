```python
def price_clean(price):
    price = price.replace(",", "").replace("$","")
    if price != "":
        return int(price)
    else:
        return price
    
def worthit_clean(worthit):
    if (worthit != 'Worth It') and (worthit != 'Not Worth It') and (worthit != 'Not Sure'):
        return ''
    else:
        return worthit
```

## LDA

```python
nltk_stopwords = stopwords.words('english')+["did", "inner", "outer", "arms","treatment", "procedure", "I", "done", "coolsculpting", "area", "day", 'fat', "cool sculpting", "pain", "results", "get", "would", "sculpting", "like", "abdomen","handles", "stomach", "lower", "love", "thighs", "legs","per", "cool","made", "went", "go", "coolsculpt", "belly", "still", "little","dont", "tummy"]
```

```python
# Create a CountVectorizer for parsing/counting words
count_vectorizer = CountVectorizer(ngram_range=(1, 3),  
                                   stop_words = nltk_stopwords, token_pattern="\\b[a-z][a-z]+\\b", max_df=0.95, min_df = 0.01)
count_vectorizer.fit(data)
```

```python
# Create the term-document matrix
# Transpose it so the terms are the rows
counts = count_vectorizer.transform(data).transpose()
```

```python
# Convert sparse matrix of counts to a gensim corpus
corpus = matutils.Sparse2Corpus(counts)
```

```python
id2word = dict((v, k) for k, v in count_vectorizer.vocabulary_.items())
```

```python
dict((v, k) for k, v in count_vectorizer.vocabulary_.items())
```

```python
teste = [k for k in count_vectorizer.vocabulary_.items()]
```

```python
teste_lst = []
for i in teste:
    teste_lst.append(i[0])
    
teste_lst
```

```python
# Create lda model (equivalent to "fit" in sklearn)
lda = models.LdaModel(corpus=corpus, num_topics=6, id2word=id2word, passes=10)
```

```python
lda.print_topics()
```

```python
def build_texts(f):
    for line in f:
        yield gensim.utils.simple_preprocess(line, deacc=True, min_len=3)
```

```python
train_texts = list(build_texts(teste_lst))
```

```python
def process_texts(texts):
    """
    Function to process texts. Following are the steps we take:
    
    1. Stopword Removal.
    2. Collocation detection.
    3. Lemmatization (not stem since stemming can reduce the interpretability).
    
    Parameters:
    ----------
    texts: Tokenized texts.
    
    Returns:
    -------
    texts: Pre-processed tokenized texts.
    """
    texts = [[word for word in line if word not in nltk_stopwords] for line in texts]
    #texts = [bigram[line] for line in texts]
    #texts = [[word[0].split("/") for word in lemmatize(' '.join(line), allowed_tags=re.compile('(NN)'), min_length=3)] for line in texts]
    #texts = [lemmatize(' '.join(line), allowed_tags=re.compile('(NN)'), min_length=3) for line in texts]
    return (texts)
    
```

```python
train_texts = process_texts(train_texts)
```

```python
dictionary = Dictionary(train_texts)
```
