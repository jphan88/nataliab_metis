```python
import pickle
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
import spacy

%matplotlib inline
```

```python
nlp = spacy.load('en')
```

### Import databases

```python
file = open("reviews.pickle",'rb')
reviews = pickle.load(file)
file.close()
```

```python
file = open("reviews_posts.pickle",'rb')
reviews_per_post = pickle.load(file)
file.close()
```

```python
def price_clean(price):
    price = price.replace(",", "").replace("$","")
    if price != "":
        return int(price)
    else:
        return price
    
```

```python
reviews['price'] = reviews['price'].apply(price_clean)
```

```python
prices = [prices for prices in reviews[reviews['price'] != '']['price']]
```

```python
print("treatment: {}".format(reviews['treatment'][0]))
print("number of reviews: {}".format(len(reviews)))
print("number of unique users: {}".format(reviews.username.nunique()))
print("number of unique doctors: {}".format(reviews.doctor.nunique()))
print("Average Price: {}{}".format(round(np.mean(prices),2), ' dollars'))
```

### Data Cleaning

```python
# clean reviews
text_clean = []
punctuation = re.compile(r"[!#$%&\'\"()\*\+,-\./:;<=>\?@\[\\\]^_`{\|}~]")

for review in reviews_per_post.body:
    temp = review.lower()
    temp = re.sub(punctuation, " ", temp)
    text_clean.append(temp)
    
len(text_clean)
```

```python
# remove stopwords and apply lemmatization
def remove_stopwords_lemma(review):
    remain = ''
    doc = nlp(review)
    for token in doc:
        if token.is_stop:
            continue
        elif token.is_space:
            continue
        elif len(token)<=1:
            continue
        else:
            remain += token.lemma_ + ' '
    return remain


bag_of_words = []
for review in text_clean:
    bag_of_words.append(remove_stopwords_lemma(review))

len(bag_of_words)
```
