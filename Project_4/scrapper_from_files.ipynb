```python
def get_page(html):
    soup = BeautifulSoup(html, "lxml")
    return soup

def read_file(file_name):
    file = open(file_name, 'r')
    return file.read()
```

```python
def get_posts_on_page(page):
    posts = page.find('div', {'id': 'review-entries'})
    posts = posts.find_all('div', id=lambda x: x and x.startswith('entry'))
    
    posts_count = 0
    posts_return = []
    
    for post in posts: 
        
        is_post = True
        
        try:
            post_title = post.find('h3', class_='Content-title').get_text().strip()
            post_body = post.find('div', class_="Content-bodyText")
            post_expanded_body = post_body.find('div', class_='text-truncator-expanded-text')
            if(post_expanded_body):
                post_body = post_expanded_body
            post_body = post_body.find('div').get_text().strip()
        except:
            is_post = False
            
        try:
            post_number_of_likes = post.find(class_="u-marginMedium")
            post_number_of_likes = post_number_of_likes['rs-static-data'].split('\"')[3].strip()
        except:
            post_number_of_likes = "0"

        try:
            post_tags = [tag.get_text().strip() for tag in post.find_all('a', class_='Link--tag')]

        except:
            post_tags = ""
        
        if (is_post):
            post_temp = {}
            post_temp["title"]=post_title
            post_temp["body"]=post_body
            post_temp["likes"]=post_number_of_likes
            post_temp["tags"]=post_tags
            posts_return.append(post_temp)
            posts_count += 1
    
    if (posts_count > 0):
        return posts_return
    else:
        return None
    
def get_review_details(page):

    is_review = True
    review_return = {}
    
    try:
        review_breadcrumbs = page.find('ul', class_='PageHeader-breadcrumb')
        review_breadcrumbs = [tag.get_text().strip() for tag in review_breadcrumbs.find_all('span')]        
    except:
        review_breadcrumbs = []

    try:
        review_title = page.find('h1', class_='Content-title').get_text().strip()
    except:
        is_review = False
    
    try:      
        review_content = page.find('div', class_='Content')
        review_content = review_content.find('ul', class_='Content-byline')
        review_content = review_content.find_all('li', class_='Byline-item')
    except:
        is_review = False
    
    index = 1
    try:
        review_username = review_content[0].get_text().strip()
    except:
        review_username = ''
    
    try:   
        review_worthit_temp = review_content[index].get_text().strip()
        if(review_worthit_temp.find('Worth') or review_worthit_temp.find('Sure')):
            review_worthit = review_worthit_temp
            index = index + 1
        else:
            review_worthit = ''
        
    except:
        review_worthit = ''

    try:
        review_price_temp = review_content[index].get_text().strip()
        if(review_price_temp.find('$')>-1):
            review_price = review_price_temp
            index = index + 1
        else:
            review_price = ''
        
    except:
        review_price = ''
    
    try:
        review_doctor_and_location_temp = review_content[index].get_text().strip().split(',')
        review_doctor = ''
        review_location = ''

        #if more than 1 comma, then it have both location and doctor's name
        review_dal_temp_len = len(review_doctor_and_location_temp)
        if (review_dal_temp_len>2):
            review_location = review_doctor_and_location_temp[review_dal_temp_len-2:review_dal_temp_len]
            review_location = ','.join(review_location).strip()
            review_doctor = review_doctor_and_location_temp[0:review_dal_temp_len-2]
            review_doctor = ','.join(review_doctor).strip()
        
        #if it has 2 commas, then it only have the location (Seattle,WA)
        if (review_dal_temp_len==2):
            review_location = review_doctor_and_location_temp[review_dal_temp_len-2:review_dal_temp_len]
            review_location = ','.join(review_location).strip()
            review_doctor = ''
    except:
        review_doctor = ''
        review_location = ''


    
    if (is_review):
        review_return['title']=review_title
        review_return['breadcrumbs']=review_breadcrumbs
        review_return['username']=review_username
        review_return['worthit']=review_worthit
        review_return['price']=review_price
        review_return['doctor']=review_doctor
        review_return['location']=review_location
        
        posts = get_posts_on_page(page)
        
        if(posts):
            review_return['posts'] = posts
        else:
            return None
        
        return review_return
    else:
        return None

def get_review_on_page_debug(page):
    review = get_review_details(page)
    
    if(review):
        print("Title: {}".format(review['title']))
        print("Breadcrumbs: {}".format(review['breadcrumbs']))
        print("Username: {}".format(review['username']))
        print("Worth it: {}".format(review['worthit']))
        print("Price: {}".format(review['price']))
        print("Doctor: {}".format(review['doctor']))
        print("Location: {}".format(review['location']))
        
        posts = review['posts']
        print("Posts:")
        for post in posts:
            print("Title: {}".format(post['title']))
            print("Body: {}".format(post['body']))
            print("Likes: {}".format(post['likes']))
            print("Tags: {}".format(post['tags']))
    
        return review
    else:
        print("Not a procedure review")
        return None

    
def get_review_on_page(page):
    review = get_review_details(page)
    
    if(review):
        posts = review['posts']
        print("A procedure review")
        return review
    else:
        print("Not a procedure review")
        return None
```
