```python
import pandas as pd
import matplotlib.pyplot as plt
import reverse_geocoder as rg
import censusgeocode as cg
import requests
import json
%matplotlib inline


dados = pd.read_csv(".\python\df_MTA.txt")
```

```python
stations = dados.groupby(["STATION"], sort=False, as_index=False)["geo"].first()
stations[:3]
```

```python
resp = requests.get('http://data.fcc.gov/api/block/find?format=json&latitude=40.6508606878022&longitude=-73.94945514035334&showall=false')
resp.json()['Block']['FIPS'][:11]
```

```python
def latlong(geo):
    lat = geo[1:].split(',')[0]
    long = geo[:-1].split(', ')[1]
    url = 'http://data.fcc.gov/api/block/find?format=json&latitude='+lat+'&longitude='+long+'&showall=false'
    resp = requests.get(url)
    if resp.status_code != 200:
        result = 0
    else: 
        result = resp.json()['Block']['FIPS'][:11]
    return(result)

latlong("(40.6508606878022, -73.94945514035334)")
```

```python
#takes ~20min or more to geocode
stationsbkp = stations
stations['CensusTract'] = stations['geo'].apply(latlong)
stations
```

```python
stations.to_csv(".\python\stations.csv")
#stations = pd.read_csv(".\python\stations.csv", index_col='Unnamed: 0', dtype=object)

```

```python
demographics = pd.read_csv("./python/nyc_census_tracts.csv")
stations = stationsbkp
stations['CensusTract'] = stations['censustract']

stations['CensusTract'] = pd.to_numeric(stations['CensusTract'])
demographics['CensusTract'] = pd.to_numeric(demographics['CensusTract'])

demographics[:10]
```
