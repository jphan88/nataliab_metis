```python
cat_data = pd.get_dummies(df_math[['school', 'sex','address', 'famsize', 'Pstatus', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']],drop_first=True)
nominal_cat = pd.get_dummies(df_math[['Mjob','Fjob','reason','guardian']],drop_first=True)
numeric_data = df_math[['age','Medu','Fedu','traveltime','studytime','famrel','freetime','goout','Dalc','Walc','health','absences','failures']]
```

```python
grades = pd.DataFrame()
#pass = 0, fail = 1
def passfail(grade):
    if grade >=10:
        return 0
    if grade <10:
        return 1
grades['G1'] = df_math['G1'].apply(passfail)
grades['G2'] = df_math['G2'].apply(passfail)
grades['G3'] = df_math['G3'].apply(passfail)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='G3',data=grades,palette='RdBu_r')
```

```python
df = pd.concat([grades, cat_data,numeric_data, nominal_cat], axis = 1)
df.head()
```

```python
df.corr()
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.corr(), cmap = 'coolwarm')
```

```python
X = df.drop(['G1','G2','G3'], axis = 1)
y = df['G1']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

## Logistic Regression

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
y_pred = logmodel.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    #if normalize:
    #    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    #    print("Normalized confusion matrix")
    #else:
    #    print('Confusion matrix, without normalization')

    #print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# generate evaluation metrics
print ("Train - Accuracy : {:.2f}%".format(metrics.accuracy_score(y_train, logmodel.predict(X_train))*100))
print ("Test - Accuracy : {:.2f}%".format(metrics.accuracy_score(y_test, logmodel.predict(X_test))*100))
```

## Using KNN



```python
normalizer = preprocessing.Normalizer()
X_norm = normalizer.fit_transform(X)
```

```python
X_train_norm, X_test_norm, y_train, y_test = train_test_split(X_norm, y, test_size=0.33, random_state=42)
```

```python
#y score is a list of number...
k_list = [1, 2, 5, 10, 20, 30, 50]
k_x = []
k_y = []
for k in k_list:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train_norm,y_train)
    y_scores_knn = knn.predict_proba(X_test_norm)[:,1]
    k_x.append(k)
    k_y.append(roc_auc_score(y_test, y_scores_knn))

plt.plot(k_x, k_y, linestyle='dashed', marker='o')
plt.rcParams['font.size'] = 12
plt.title('AUC vs. k neighbors')
plt.xlabel('k')
plt.ylabel('AUC')
```

k = 10

```python
knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train_norm,y_train)
y_pred_knn = knn.predict(X_test_norm)
plot_confusion_matrix(confusion_matrix(y_test,y_pred_knn), ['pass', 'fail'])
```

```python
print(classification_report(y_test,y_pred_knn))
```

```python
# generate evaluation metrics
print ("Train - Accuracy : {:.2f}%".format(metrics.accuracy_score(y_train, knn.predict(X_train))*100))
print ("Test - Accuracy : {:.2f}%".format(metrics.accuracy_score(y_test, knn.predict(X_test))*100))
```

## Decision Trees

```python
max_depth_list = [1,2,5, 10, 15, 20, 25]
max_depth_x = []
max_depth_y = []
for k in max_depth_list:
    dtree = DecisionTreeClassifier(max_depth=k)
    dtree.fit(X_train,y_train)
    y_scores_dtree = dtree.predict_proba(X_test)[:,1]
    max_depth_x.append(k)
    max_depth_y.append(roc_auc_score(y_test, y_scores_dtree))

#roc_auc_score vs. k

plt.plot(max_depth_x, max_depth_y, linestyle='dashed', marker='o')
plt.rcParams['font.size'] = 12
plt.title('AUC vs. Max Depth')
plt.xlabel('Max Depth')
plt.ylabel('AUC')
```

```python
plot_confusion_matrix(confusion_matrix(y_test,y_pred_dtree), ['pass', 'fail'])
```

```python
# generate evaluation metrics
print ("Train - Accuracy : {:.2f}%".format(metrics.accuracy_score(y_train, dtree.predict(X_train))*100))
print ("Test - Accuracy : {:.2f}%".format(metrics.accuracy_score(y_test, dtree.predict(X_test))*100))
```

## Random Forest

```python
n_estimators_list = [5, 10, 15, 20, 25, 50,100,150]
n_estimators_x = []
n_estimators_y = []
for k in n_estimators_list:
    rfc = RandomForestClassifier(n_estimators=k)
    rfc.fit(X_train, y_train)
    y_scores_rfc = rfc.predict_proba(X_test)[:,1]
    n_estimators_x.append(k)
    n_estimators_y.append(roc_auc_score(y_test, y_scores_rfc))

plt.plot(n_estimators_x, n_estimators_y, linestyle='dashed', marker='o')
plt.rcParams['font.size'] = 12
plt.title('AUC vs. n Estimators')
plt.xlabel('n')
plt.ylabel('AUC')
```

### ROC Curve

```python
y_scores_knn = knn.predict_proba(X_test_norm)[:,1]
fpr_knn, tpr_knn, thresholds_knn = roc_curve(y_test, y_scores_knn)

y_scores_log = logCV.predict_proba(X_test)[:,1]
fpr_log, tpr_log, thresholds_log = roc_curve(y_test, y_scores_log)

y_scores_dtree = dtree.predict_proba(X_test)[:,1]
fpr_dtree, tpr_dtree, thresholds_dtree = roc_curve(y_test, y_scores_dtree)

y_scores_rfc = rfc.predict_proba(X_test)[:,1]
fpr_rfc, tpr_rfc, thresholds_rfc = roc_curve(y_test, y_scores_rfc)

plt.figure(figsize = (12,6))
plt.plot(fpr_knn, tpr_knn, label='KNN')
plt.plot(fpr_log, tpr_log, label='Logistic Regression')
plt.plot(fpr_dtree, tpr_dtree, label='Decision Tree')
plt.plot(fpr_rfc, tpr_rfc, label='Random Forest')

# Legend with some customizations.
legend = plt.legend(loc='lower right', shadow=True)
frame = legend.get_frame()
frame.set_facecolor('0.90')
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(3)  # the legend line width


plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.rcParams['font.size'] = 12
plt.title('ROC curve for Students Performance Classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')

plt.grid(True)
```
